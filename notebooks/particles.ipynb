```python
import matplotlib.pyplot as plt
import scienceplots  # noqa: F401
import numpy as np
import pandas as pd
plt.style.use(["science", "ieee", "notebook"])

plt.rcParams["font.size"] = 12
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["figure.figsize"] = (9, 4)
```

```python
def get_distribution(sheet_name: str) -> tuple[float,float]:
    """
    Function to evaluate the Sauter Mean Diameter (SMD) of the given
    experiment under `data/particles.xlsx`.

    Returns
    -------
    dsp: float
        The SMD in micra (μm)

    rho_s: float
        Solid particle density (g/cm3)
    """
    data = pd.read_excel("data/particles.xlsx", sheet_name=sheet_name)
    
    # manual logic to obtain the density, kinda dumb
    if sheet_name.startswith("VIDRO"):
        rho_s = 1.50 
    elif sheet_name.startswith("PAINCO"):
        rho_s = 1.17 
        
    xi, Di = data.iloc[:, 2], data.iloc[:, 6]
    Dim, Dip = data.iloc[:, 4], data.iloc[:, 5]
    dsp = sum(xi / 100) / (sum(xi / 100 / Di))
    print(dsp)

    plt.figure()
    plt.yticks(np.arange(0, 110, 10))
    plt.grid(alpha=0.3)
    plt.bar(Di, xi, width=Dim - Dip, fc="#45474B", ec="k")
    plt.axvline(dsp, label=r"d$_{sp}$", c="k")
    plt.legend(ncol=2)
    plt.xlabel("Diâmetro (μm)")
    plt.ylabel("Frequência (%100)")
    plt.savefig(
        f"images/particle-distribution-{sheet_name}.pdf", bbox_inches="tight", dpi=300
    )
    plt.show()

    return (dsp, rho_s)
```

```python
particle1 = get_distribution("VIDRO-B2")
particle2 = get_distribution("VIDRO-B3")
particle3 = get_distribution("PAINCO")

particles = [particle1, particle2, particle3]
```

```python
# plot geldart curve and insert the particles used in it
d_arange = np.arange(0.1, 2600, 0.01)  # particle size in μm

plt.figure(figsize=(9, 4))
plt.ylabel(r"($\rho_s - \rho_f$) (g/cm³)")
plt.xlabel(r"d$_{sp}$ (μm)")
plt.xlim(30, 2600)
plt.ylim(0.01, 7.0)
plt.xscale("log", base=10)

default_alpha = 0.10
# GRUPO A:
# (ps - pf)d' <= 225
rho_diff_A = 225 / d_arange
plt.plot(d_arange, rho_diff_A, "k-")
plt.fill_between(d_arange, 0, rho_diff_A, alpha=default_alpha, color="lime")

# Grupo D:
# (ps - pf)(d')^2 >= 10^6
rho_diff_D = 1e6 / (d_arange**2)
plt.plot(d_arange, rho_diff_D, "k-")
plt.fill_between(
    d_arange,
    rho_diff_A,
    rho_diff_D,
    where=(rho_diff_A < rho_diff_D),
    alpha=default_alpha,
    color="blue",
)
plt.fill_between(d_arange, rho_diff_D, 8, alpha=default_alpha, color="red")

# Add group labels as text
plt.text(35, 3.5, "Grupo A", ha="left", va="center", weight="bold")
plt.text(150, 4.75, "Grupo B", ha="center", va="center", weight="bold")
plt.text(1500, 5.0, "Grupo D", ha="right", va="center", weight="bold")

for dnum, particle in enumerate(particles):
    dsp, rho_s = particle[0], particle[1]
    plt.scatter(dsp, rho_s - 0.001225, marker="s", c="k", s=30)
    plt.text(dsp, rho_s+0.2, f"({dnum + 1})", ha="center", weight="bold")

plt.savefig("images/particles-geldart.pdf", dpi=300, bbox_inches="tight")
plt.show()
```
